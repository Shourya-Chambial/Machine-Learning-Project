# DF5 - 16000 approx

```python
# http://www.aclweb.org/anthology/N16-2013
labeled_ids = pd.read_csv('NAACL_SRW_2016.csv', names = ['id', 'label'])

labeled_ids.head(2)
```

```python
import tweepy
```

```python
consumer_key = ''
consumer_secret = ''
access_token = ''
access_secret = ''
#cannot be revealed

# authenticate as per tweepy docs
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)


# create api object
api = tweepy.API(auth, wait_on_rate_limit = True, wait_on_rate_limit_notify = True)
tweet = api.get_status(572342978255048705, tweet_mode = 'extended')
print(tweet)
```

```python
import json
import time
i = 0
j = 0
failed_ids = []

# open file to write json objects from api
with open('tweets.txt', 'w') as outfile:
    c=0
    for _ in labeled_ids.id:
        if c%15:
          time.sleep(20)
        # try-except block since few tweet IDs in the archive may have been deleted
        try: 
            tweet = api.get_status(_, tweet_mode = 'extended')
        except:
            failed_ids.append(_)
            j = j+1
            print(f'Failed: {_} | {j} of {len(labeled_ids.id)}')
            continue
            
        
        # print the number of tweets collected
        print(f'Success: {_} | {i} of {len(labeled_ids.id)}')
        i = i+1
        
        # dump the json object corresponding to the tweet collected from the api
        json.dump(tweet._json, outfile)
        outfile.write('\n')
        c+=1
print(f'Number of Successful Tweets Querried: {i}')
print(f'Number of Failed Queries: {j}')
```

```python
data = []
with open('tweets.txt') as f:    
        for line in f:         
            data.append(json.loads(line))
# print(data)
df_api = pd.DataFrame(data)

#select columns of interest
columns_of_interest = ['id', 'full_text']
df_api = df_api[columns_of_interest]
```

```python
# join the dataframes with ID's and tweets
df5 = labeled_ids.merge(df_api, left_on = 'id', right_on = 'id', how = 'left');

# drop the id's whose tweets could not be retrieved
df5.dropna(how = 'any', inplace = True)
```

```python
df5.to_csv('labeled_tweets.csv', index = None)
```

```python
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
corpus = []
for i in range(0, len(df.index)):
  # print(df['full_text'][i])
  tweet = str(df['full_text'][i])
  tweet = tweet.lower()
  tweet = tweet.split()
  ps = PorterStemmer()
  all_stopwords = stopwords.words('english')
  all_stopwords.remove('not')
  tweet = [ps.stem(word) for word in tweet if not word in set(all_stopwords)]
  tweet = ' '.join(tweet)
  df['full_text'][i]=tweet
  corpus.append(tweet)
```
